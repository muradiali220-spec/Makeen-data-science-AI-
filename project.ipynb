```python
df["Hours_of_freetime"]=+df['Extracurricular_Hours_Per_Day']+df['Social_Hours_Per_Day']+df['Physical_Activity_Hours_Per_Day']
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Randomly select 200 rows from your DataFrame
df= df.sample(n=200, random_state=42)

plt.plot(df['Study_Hours_Per_Day'],df['GPA'],marker='o')
plt.title("Student hours VS GPA")
plt.xlabel("Hours OF STUDY")
plt.ylabel("GPA")
plt.grid(True)
plt.show()
```

```python
df= df.sample(200)  # plot only 500 random points
plt.scatter(df['Hours_of_freetime'], df['GPA'], alpha=0.5)
plt.title('Student hours')
plt.xlabel('Freetime')
plt.ylabel('GPA')
plt.show()
```

```python
# Randomly select 100 rows from your DataFrame
df_sample = df.sample(n=200, random_state=42)

# Then plot it
plt.scatter(df_sample['Study_Hours_Per_Day'], df_sample['GPA'], alpha=0.5)
plt.title('Student hours vs GPA')
plt.xlabel('Study Hours Per_Day')
plt.ylabel('GPA')
plt.show()
```

```python

# Histogram for Python marks
plt.hist(df["GPA"], bins=10, color='skyblue', edgecolor='black')
plt.title("GPA FOR STUDENTS")
plt.xlabel("GPA")
plt.ylabel("Number of Students")
plt.locator_params(axis='y',integer=True)# number stu int
plt.show()
```

```python
plt.scatter(df['Stress_Level'],df['GPA'],color='blue',marker='^',alpha=0.5)
plt.title("STUDENT GPA")
plt.xlabel("Stress_Level")
plt.ylabel("GPA")
plt.show()
```

```python
# Pie chart for student distribution by stress
df["Stress_Level"].value_counts().plot.pie(autopct='%1.1f%%', startangle=0)
plt.title("Students Stress")
plt.ylabel("")  # Hide default y-label
plt.show()
```

```python
#bar plot 
plt.bar(df['Stress_Level'],df['Hours_of_freetime'],color='red',width=0.8)#can user o or : for marker
plt.title("STRESS vs FREETIME")
plt.xlabel("Stress_Level")
plt.ylabel("Hours_of_freetime")
plt.show()
```
